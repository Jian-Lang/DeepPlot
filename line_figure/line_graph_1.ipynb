This is the first style of line graph. Firstly, import essential libs and set a font style.

```python

%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Times New Roman'

```

Then, define plotting data.

```python
x = ['0', '1', '2', '6', '11']
y = [1.7154, 1.7064, 1.7112, 1.7315, 1.7280]
```

Create plot:

```python
fig, ax = plt.subplots(figsize=(6, 3.5))

ax.plot(x, y, marker='p', color='black', linewidth=1, markersize=16, markerfacecolor='#f45f74')
```

Set axis limits and ticks:

```python
ax.set_ylim(1.7, 1.74)
ax.set_yticks(np.arange(1.70, 1.74, 0.01))
ax.set_yticklabels(['1.70', '1.71', '1.72', '1.73', '1.74'], fontsize=20)
ax.set_xticklabels(x, fontsize=20)
```

Add grid to figure:

```python
ax.grid(True, linestyle='--', alpha=0.7)
```

Set top and right spines invisiable to make the plot nicer.

```python

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

If u need to draw a line to indicate the min or max value point, using the follow codes:

```python
min_value = min(y)
min_index = y.index(min_value)
ax.axvline(x=min_index, color='grey', linestyle='--', linewidth=2)
```

Set axis labels:

```python
ax.set_ylabel('Y', fontsize=22)
ax.set_xlabel('X', fontsize=22)
```

Display ur plot with tight layout:

```python
# %matplotlib inline
plt.tight_layout()
# out.show()  
plt.show(block=False)
```
